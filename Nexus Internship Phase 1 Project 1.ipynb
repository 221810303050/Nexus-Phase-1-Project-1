```python
conversation_context = []

def remember_interaction(user_input, bot_response):
    conversation_context.append((user_input, bot_response))

def recall_context():
    return conversation_context

def greet_user():
    return "Hello! How can I help you today?"

def farewell_message():
    return "Goodbye! Have a great day!"

def ask_user_custom(custom_questions=None):
    default_questions = [
        "What is your name?",
        "How can I assist you today?",
        "Is there anything else you need help with?"
    ]
    #if custom_questions:
    #    questions = default_questions + [q for q in custom_questions if q not in default_questions]
    #else:
    #    questions = default_questions
    #return questions
    questions = list(default_questions)
    
    added_custom_questions = False
    
    for q in custom_questions:
        if q not in questions and custom_questions:
            questions.append(q)
            added_custom_questions = True
    return questions

def respond_to_basic_questions(user_input):
    responses = {
        "what is your name?": "I am a chatbot created to assist you.",
        "how are you?": "I'm just a program, but I'm here to help you!",
        "what can you do?": "I can answer basic questions and remember some context.",
        "where are you from?": "I exist in the digital world.",
        "who created you?": "I was created by a developer as part of an internship project."
    }
    return responses.get(user_input.lower(), "I'm sorry, I don't understand that question.")

def handle_error():
    return "I'm sorry, I didn't understand that. Could you please rephrase?"

def process_user_responses(question, user_response):
    responses = {
        "what is your name?": "Nice to meet you!",
        "how can i assist you today?": "I'm here to help you with any questions you have.",
        "is there anything else you need help with?": "Sure, what else do you need?",
        "do you need any specific information?": "I'm here to provide any specific information you need.",
        "are you looking for help with something in particular?": "Yes, let me know what specific help you need."
    }
    return responses.get(question.lower(), "I'm here to help with anything you need.")

def main():
    print(f"Bot: {greet_user()}")
    
    questions = ask_user_custom(custom_questions=[
        "Do you need any specific information?",
        "Are you looking for help with something in particular?"
    ])
    
    for question in questions:
        user_response = input(f"Bot: {question}\nYou: ")
        response = process_user_responses(question, user_response)
        remember_interaction(question, response)
        print(f"Bot: {response}")
    
    while True:
        user_input = input("You: ")
        if user_input.lower() in ["bye", "exit", "quit"]:
            print(f"Bot: {farewell_message()}")
            break
        #elif any(q in user_input.lower() for q in ["what is your name", "how are you", "what can you do", "where are you from", "who created you"]):
        #    response = respond_to_basic_questions(user_input)
        else:
            response = handle_error()
        remember_interaction(user_input, response)
        print(f"Bot: {response}")
        
        context = recall_context()
        print("Bot: Here's what we've discussed so far:")
        for interaction in context:
            print(f"You: {interaction[0]} -> Bot: {interaction[1]}")

if __name__ == "__main__":
    main()
```

## Objective:
1. Conversation Context Management:
To enable the chatbot to remember previous interactions, ensuring it can provide responses that take into account the ongoing conversation.

Implementation:
The conversation_context list stores pairs of user inputs and the bot’s replies. This list serves as the chatbot's memory, allowing it to refer back to earlier exchanges during the conversation.

2. Modular Design:
To structure the chatbot’s functionality into separate, reusable functions, making the code more organized and easier to manage.

Implementation:
The chatbot's features are broken down into distinct functions:

remember_interaction(): Saves the exchanges between the user and the bot.
recall_context(): Retrieves the saved interactions.
greet_user() and farewell_message(): Provide standard greetings and goodbyes.
ask_user_custom(): Generates and customizes questions for the user.
respond_to_basic_questions(): Handles simple, predefined queries.
handle_error(): Deals with inputs the bot doesn’t understand.
process_user_responses(): Processes and responds to the user’s answers to specific questions.

3. Customizable User Interaction:
Objective:
To create a flexible interaction model where the bot can ask both default and custom questions.
Implementation:
The ask_user_custom() function begins with a list of default questions and allows for the addition of custom questions. This approach ensures that the bot can handle a variety of interaction scenarios, making it adaptable to different user needs.

4. Error Handling:
Objective:
To manage cases where the bot doesn’t understand the user’s input and to provide a smooth user experience even when errors occur.
Implementation:
The handle_error() function is designed to provide a generic response when the bot encounters an unfamiliar input. This ensures that the bot doesn't break or give a confusing response when it can't understand the user's request.

5. Conversation Summary:
Objective:
To provide users with a summary of the entire conversation, reinforcing the chatbot’s memory feature and enhancing transparency in the interaction.
Implementation:
After each interaction, the bot prints out a summary of the conversation so far, showing both the user’s questions and the bot’s responses. This helps users keep track of the conversation and provides a clear record of the interaction.

6. Main Interaction Loop:
Objective:
To maintain an ongoing conversation with the user, allowing them to interact with the bot until they choose to exit.
Implementation:
The main() function contains a loop that continues to prompt the user for input and responds until the user types an exit command. This loop ensures that the bot remains active and responsive throughout the session.
